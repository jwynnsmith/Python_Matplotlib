```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
got_city_df = pd.read_csv(city_data_to_load)
got_ride_df = pd.read_csv(ride_data_to_load)
```

```python
# Combine the data into a single dataset
main_df = pd.merge(got_city_df, got_ride_df, on='city')

# Display the data table for preview

main_df.head()
```

## Bubble Plot of Ride Sharing Data

## Total Fares by City Type

```python
# Calculate Type Percents
type_df = main_df.groupby(["type"]) 
tot_fare_per_type = type_df["fare"].sum()
print(tot_fare_per_type[0])
print(tot_fare_per_type)
tot_fare_all_type = tot_fare_per_type.sum()
print(tot_fare_all_type)


# Build Pie Chart
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [tot_fare_per_type[0], tot_fare_per_type[1], tot_fare_per_type[2]]

# The colors of each section of the pie chart
colors = ["yellow", "lightblue", "coral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% Total Fares by City Type")

# Save Figure
plt.savefig("City_Type_Fares.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
type_df = main_df.groupby(["type"]) 
tot_rides_per_type = type_df["city"].count()
print(tot_rides_per_type[0])
print(tot_rides_per_type)

# Build Pie Chart

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [tot_rides_per_type[0], tot_rides_per_type[1], tot_rides_per_type[2]]

# The colors of each section of the pie chart
colors = ["yellow", "lightblue", "coral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% Total Rides by City Type")

# Save Figure
plt.savefig("City_Type_Rides.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
type_df = main_df.groupby(["type"]) 
tot_driver_per_type = type_df["driver_count"].sum()
print(tot_driver_per_type[0])
print(tot_driver_per_type)

# Build Pie Charts

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [tot_driver_per_type[0], tot_driver_per_type[1], tot_driver_per_type[2]]

# The colors of each section of the pie chart
colors = ["yellow", "lightblue", "coral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% Total Driver Count by City Type")

# Save Figure
plt.savefig("City_Type_Driver_Ct.png")

# Show Figure
plt.show()
```
